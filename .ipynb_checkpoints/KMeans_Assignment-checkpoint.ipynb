## K Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

## Fit the model to all the data except for the Private label.

```python
target = df.drop('Private',axis=1)
kmeans.fit(target)
```

## What are the cluster center vectors?

```python
kmeans.cluster_centers_
```

## Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
df['Cluster'] = df['Private'].replace(to_replace=['Yes','No'],value=[1,0])
# df = df['Private'].replace(to_replace='No',value=0) ignore
```

## Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
kmeans.labels_
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(kmeans.labels_, df['Cluster'])
```

```python
print(classification_report(kmeans.labels_,df['Cluster']))
```
